```python
#Rock paper and scissors 

win1 = 0
win2 = 0
while True:
    p1 = input("P1 : R,P,S")

    if (p1 != 'R'and p1 != 'r' and p1 != 'P' and p1!= 'p' and p1!= 'S' and p1!= 's'):
        print("choose valid")
        break

    p2 = input("P2 : R,P,S")

    if (p2 != 'R'and p2 != 'r' and p2 != 'P' and p2!= 'p' and p2!= 'S' and p2!= 's'):
        print("choose valid")
        break

    if (p1 == "R" or p1 == 'r') and (p2 == "P" or p2 == 'p'):
        print("p2 wins")
        win2 += 1
    elif (p1 == "P" or p1 == 'p') and (p2 == "R" or p2 == 'r'):
        print("p1 wins")
        win1 += 1
    elif (p1 == "P" or p1 == 'p') and (p2 == "S" or p2 == 's'):
        print("p2 wins")
        win2 += 1
    elif (p1 == "S" or p1 == 's') and (p2 == "P" or p2 == 'p'):
        print("p1 wins")
        win1 += 1
    elif (p1 == "S" or p2 == 's') and (p2 == "R" or p2 == 'r'):
        print("p2 wins")
        win2 += 1
    elif (p1 == "R" or p1 == 'r') and (p2 == "S" or p2 == 's'):
        print("p1 wins")
        win1 += 1
    else:
        print("Choose Valid")
        continue

    n = input("do you want to continue the game? Y/N")
    if n == 'Y' or n == 'y':
        continue
    else:
        break

print("score of p1 =", win1)
print("score of p2 =", win2)

if win1 == win2:
    print("tie")
elif max(win1,win2) == win1:
    print("P1 is the winner")
else:
    print("p2 is the winner")
```

```python
#Rock paper and scissors

win1 = 0
win2 = 0

while True:

    while True:
        p1 = input("P1: R, P, S ").lower()
        if p1 in ['r', 'p', 's']:
            break
        print("Invalid choice! Choose R, P, or S.")

    while True:
        p2 = input("P2: R, P, S ").lower()
        if p2 in ['r', 'p', 's']:
            break
        print("Invalid choice! Choose R, P, or S.")


    if p1 == p2:
        print("It's a tie!")

    elif (p1 == 'r' and p2 == 's') or (p1 == 'p' and p2 == 'r') or (p1 == 's' and p2 == 'p'):
        print("P1 wins this round!")
        win1 += 1
    else:
        print("P2 wins this round!")
        win2 += 1

    while True:
        n = input("Do you want to continue? Y/N: ").lower()
        if n in ['y', 'n']:
            break
        print("Please enter Y or N.")

    if n == 'n':
        break


print("\nFinal Score")
print("Score of P1 =", win1)
print("Score of P2 =", win2)

if win1 == win2:
    print("Match Tie")
elif win1 > win2:
    print("P1 is the winner ")
else:
    print("P2 is the winner ")
```

```python
while True:
    print("select from the following options :")
    print("1. Lowercase to uppercase \n 2. Uppercase to lowercase . \n 3. Exit")
    n = input("")
    while True:
        if n != '1' or n != '2' or n != '3':
            print("Enter a valid option")
        else:
            break
    

    if n == '1':
        print("You have chosen Lowercase to uppercase")

        while True:
            c =input("Enter a lowercase caracter")
            if c.islower() and len(c) == 1:
                print(c.upper())
                break
            print("Enter a valid lowercase character")
        
    elif n == '2':
        print("You have chosen Uppercase to Lowercase")
        
        while True:
            c =input("Enter an Uppercase caracter")
            if c == c.upper() and len(c) == 1:
                print(c.lower())
                break
            print("Enter a valid Uppercase character")
    
    elif n == '3':
        print("Exiting the program")
        break

    else:
        print("Enter the correct Option")
    
    
    while True:
        n = input("Do you want to continue? Y/N: ").lower()
        if n in ['y', 'n']:
            break
        print("Please enter Y or N.")

    if n == 'n':
        break
```
